```python
class TextDataset:

    def __init__(self, data):
        self.data = data

    def __len__(self):
        return self.data.shape[0]

    def __getitem__(self, idx):
        row = self.data.iloc[idx]

        enc = tokenizer(
            row["text"],
            max_length=10,
            truncation=True,
            padding="max_length"
        )

        return {
            "input_ids": torch.tensor(enc["input_ids"]),
            "attention_mask": torch.tensor(enc["attention_mask"]),
            # "label": torch.tensor(row["label"]),
        }
```

```python
df = pd.read_csv("/kaggle/input/imdb-dataset-of-50k-movie-reviews/IMDB Dataset.csv").rename(columns={"review": "text"})
df.head()
```

```python
id2label = {0: "negative", 1: "positive"}
label2id = {label: id_ for id_, label in id2label.items()}

df["label"] = df["sentiment"].map(label2id)

print(df.shape)
df.head()
```

```python
tokenizer = transformers.AutoTokenizer.from_pretrained("/kaggle/input/imdb-review-clf/results")

model = transformers.AutoModelForSequenceClassification.from_pretrained("/kaggle/input/imdb-review-clf/results")
```

```python
df1 = TextDataset(df)
[df1[0], df1[1]]
```

```python
dl = torch.utils.data.DataLoader(
    df1,
    batch_size=2,
    shuffle=False,
    num_workers=2,
)
```

```python
model.to("cuda")
```

```python
for idx, batch in enumerate(dl):
    print(batch)

    batch = {k: v.to('cuda') for k, v in batch.items()}

    with torch.no_grad():
        out = model(input_ids=batch["input_ids"], attention_mask=batch["attention_mask"])

    if idx == 5:
        break
```
